```python
import instaloader
import pandas as pd
import numpy as np
import os
```

Firstly I made a virtual environment named Instaloader_Project

In that I installed the instaloader

```python
df=pd.DataFrame(columns=['Number','Caption','Like'])

L = instaloader.Instaloader()

def download_post_if_image(post, num_posts):

    if post.typename =='GraphSideCar':

        for node in post.get_sidecar_nodes():
            if node.is_video:
                return False
            
            else:
                L.download_post(post, target=os.path.join(str(num_posts)))
                return True
            
    elif post.typename=='GraphImage':
        L.download_post(post, target=os.path.join(str(num_posts)))
        return True

    else:
        return False


def download_posts(username, num_posts):

    global df

    profile = instaloader.Profile.from_username(L.context, username)

    for post in profile.get_posts():

        if num_posts <= 0:
            break

        if download_post_if_image(post, num_posts):
            
            post_dict={'Number':str(num_posts), 'Caption':str(post.caption),'Like':str(post.likes), 'URL':str(post.url)}
            
            df=pd.concat([pd.DataFrame([post_dict]),df], ignore_index=True)

            print("Printing")

            num_posts -= 1

username_to_scrape = "natgeo" 
number_of_posts = 1000

download_posts(username_to_scrape, number_of_posts)

df.to_csv("data_sheet.csv")
```
